```python
import os
import wget
from pathlib import Path
import json

#phone for syntetic datasets
#mac for not syntetic datasets

datasets=[
{   'name' :'nerf_synthetic','objects':['chair','drums','ficus','hotdog','lego','materials','mic','ship'], 'device':'phone'}
]

for dataset in datasets:
    # Set the base directory
    base_dir = dataset['name']
    os.makedirs(base_dir, exist_ok=True)
    os.chdir(base_dir)

    for object in dataset['objects']:
        device=dataset['device']
        base_obj_dir = object+"_"+device

        os.makedirs(base_obj_dir, exist_ok=True)
        os.chdir(base_obj_dir)

        
        # Download mlp weights
        mlp_url = "https://storage.googleapis.com/jax3d-public/projects/mobilenerf/mobilenerf_viewer/"+object+"_"+device+"/mlp.json"
        wget.download(mlp_url)

        # Note the number of objects
        with open("mlp.json", "r") as file:
            mlp_data = json.load(file)
            num_objects = mlp_data["obj_num"]
            print(f"Number of objects: {num_objects}")

        # Download obj files
        for i in range(int(num_objects)):
            obj_url = f"https://storage.googleapis.com/jax3d-public/projects/mobilenerf/mobilenerf_viewer/{object}_{device}/shape{i}.obj"
            wget.download(obj_url)

        # Download png files
        for i in range(int(num_objects)):
            png_url = f"https://storage.googleapis.com/jax3d-public/projects/mobilenerf/mobilenerf_viewer/{object}_{device}/shape{i}.pngfeat0.png"
            wget.download(png_url)

            png_url = f"https://storage.googleapis.com/jax3d-public/projects/mobilenerf/mobilenerf_viewer/{object}_{device}/shape{i}.pngfeat1.png"
            wget.download(png_url)
        os.chdir("..")
```
